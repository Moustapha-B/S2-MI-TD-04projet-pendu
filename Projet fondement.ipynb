```python
import random
import tkinter as tk


#mot Ã  trouver
liste= ["France","Allemagne","Espagne"]

n=random.randint(0,len(liste)-1)
mot= liste[n]


#trouver le mot
def erreur():
    indice=0
    erreur=0
    while erreur<7:
        lettre=input("Entrez une lettre")
        if lettre not in mot:
            erreur+=1
        else:
            for l_du_mot in mot:
                if lettre !=l_du_mot:
                    indice+=1
                print(indice)

            #demasquer le mot a lindice 
        
    
        
print(mot)
erreur()
    
    











#interface
fenetre = tk.Tk()
fenetre.title("Pendu")
label = tk.Label(fenetre,text=len(mot)*"*")
label.pack()


fenetre.mainloop()









```
